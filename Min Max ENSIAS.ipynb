```python
#importing the libraries
import tabula
import pandas as pd
import numpy as np
```

```python
#reading the CNC results of each field
dfs_MP = tabula.read_pdf("CNC2022_MP.pdf", pages='all')
dfs_PSI = tabula.read_pdf("CNC2022_PSI.pdf", pages='all')
dfs_TSI = tabula.read_pdf("CNC2022_TSI.pdf", pages='all')
```

```python
#the TSI file mat9rach lia correctly. I had to do these steps, don't worry if you don't understand their purpose.
dftsi1=dfs_TSI.pop(0)
dftsi2=dfs_TSI.pop(0)
dfTSItest= pd.merge(dftsi1, dftsi2, on='NOM', how='outer')
dfs_TSI=[dfTSItest]+dfs_TSI
```

```python
#combining them to make dataframes
df_combined_MP = pd.concat(dfs_MP, ignore_index=True)
df_combined_PSI = pd.concat(dfs_PSI, ignore_index=True)
df_combined_TSI = pd.concat(dfs_TSI, ignore_index=True)

df_combined_PSI
```
